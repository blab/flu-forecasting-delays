```python
import altair as alt
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv(
    "results/clade_frequencies_for_h3n2.tsv",
    sep="\t",
    parse_dates=["timepoint", "future_timepoint"],
)
```

```python
df.head()
```

```python
df[(df["future_timepoint"] == "2014-07-01") & (df["delta_month"] == 6) & (df["delay_type"] == "realistic")].sort_values("absolute_forecast_error")
```

```python
df[df["frequency"] > 0.25]["clade_membership"].drop_duplicates()
```

```python
clades = [
    "A.6.8.5.2.4.1.3",
    "A.6.8.5.2.4.1.4",
]
```

```python
clade_df = df[df["clade_membership"].isin(clades)].copy()
```

```python
clade_df.head()
```

```python
clade_df["delay_type"].drop_duplicates()
```

```python
clade_df.shape
```

```python
subset_clade_df = clade_df.loc[
    (clade_df["delta_month"] == 3.0),
    [
        "delay_type",
        "clade_membership",
        "timepoint",
        "frequency",
    ]
].drop_duplicates()
```

```python
subset_clade_df.head()
```

```python
sns.relplot(
    data=subset_clade_df,
    x="timepoint",
    y="frequency",
    hue="delay_type",
    col="clade_membership",
    col_wrap=2,
    kind="line",
)

sns.despine()
```

```python
specific_clade_df = df.query("clade_membership == 'A.6.8.5.2.4.1.4'").copy()
```

```python
observed_clade_df = specific_clade_df.loc[
    (specific_clade_df["delay_type"] == "none") & (specific_clade_df["delta_month"] == 3.0),
    [
        "frequency",
        "timepoint",
    ]
].drop_duplicates()
```

```python
ideal_clade_df = specific_clade_df.loc[
    (specific_clade_df["delay_type"] == "ideal") & (specific_clade_df["delta_month"] == 3.0),
    [
        "frequency",
        "timepoint",
    ]
].drop_duplicates()
```

```python
realistic_clade_df = specific_clade_df.loc[
    (specific_clade_df["delay_type"] == "realistic") & (specific_clade_df["delta_month"] == 3.0),
    [
        "frequency",
        "timepoint",
    ]
].drop_duplicates()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 4), dpi=120)

ax = sns.lineplot(
    data=observed_clade_df,
    x="timepoint",
    y="frequency",
    ax=ax,
    label="none",
)

ax = sns.lineplot(
    data=ideal_clade_df,
    x="timepoint",
    y="frequency",
    ax=ax,
    label="ideal",
)

ax = sns.lineplot(
    data=realistic_clade_df,
    x="timepoint",
    y="frequency",
    ax=ax,
    label="realistic",
)

ax.set_xlabel("Date")
ax.set_ylabel("Proportion")

#ax.set_ylim(0, 1)

ax.legend(
    title="Delay type",
    frameon=False,
)

sns.despine()
```

```python
observed_clade_df = specific_clade_df.query("delta_month == 3.0")
```

```python
observed = alt.Chart(
    observed_clade_df
).mark_line(
    point=True
).encode(
    x="timepoint:T",
    y="frequency:Q",
    color=alt.Color("delay_type:N").sort(["none", "ideal", "realistic"]),
    tooltip=["timepoint:T", "frequency:Q", "delay_type:N"],
)

(observed).properties(
    width=800
)
```
